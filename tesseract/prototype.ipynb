```python
import os
import random
from PIL import Image

# Define the folder path
folder_path = "./out"

# Get all .jpg files in the folder
jpg_files = [file for file in os.listdir(folder_path) if file.endswith('.jpg')]

# Check if there are .jpg files in the folder
if jpg_files:
    # Select a random file
    random_file = random.choice(jpg_files)
    
    # Construct the full path
    random_file_path = os.path.join(folder_path, random_file)
    
    # Open the image
    image = Image.open(random_file_path)
    
    # Show the image
    image.show()
    
    # Print the file name
    print(f"Randomly selected file: {random_file}")
else:
    print("No .jpg files found in the specified folder.")
```

```python
# Extract the text from the image
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'
text = pytesseract.image_to_string(image)
print(text)
```
